```python
def convert_rom(value):
    value=value.replace('GB','').replace(' ','')
    if 'TB' in value:
        return int(value.replace('TB',''))*1024
    return int(value)

df['ROM']=df['ROM'].apply(convert_rom)
```

```python
df['PPI'] = ((df['resolution_width']**2 + df['resolution_height']**2)**0.5) / df['display_size']
```

```python
mapping = {
    'dual': 2,
    'quad': 4,
    'hexa': 6,
    'octa': 8
}

import re

def extract_core_count(cpu_string):
    cpu_string = str(cpu_string).lower()

    for word, count in mapping.items():
        if word in cpu_string:
            return count

    match = re.search(r'(\d+)\s*cores?', cpu_string)
    if match:
        return int(match.group(1))

    return None


df['core_count'] = df['CPU'].apply(extract_core_count).astype('float')
```

```python
def extract_thread_count(cpu_string):
    cpu_string = str(cpu_string).lower()

    for word, count in mapping.items():
        if word in cpu_string:
            return count

    match = re.search(r'(\d+)\s*threads?', cpu_string)
    if match:
        return int(match.group(1))

    return None
df['thread_count'] = df['CPU'].apply(extract_thread_count).astype('float')
```

```python
df['CPU_generation'] = df['processor'].str.extract(r'(\d+)(?:th|st|nd|rd)\s+Gen', flags=re.IGNORECASE).astype(float)
```

```python
df.loc[df['processor'].str.contains('Apple M1', na=False), 'CPU_generation'] = 1
df.loc[df['processor'].str.contains('Apple M2', na=False), 'CPU_generation'] = 2
df.loc[df['processor'].str.contains('Apple M3', na=False), 'CPU_generation'] = 3
```

```python
df['CPU_generation'].fillna(0, inplace=True)
df['core_count'].fillna(df['core_count'].mean(), inplace=True)
df['thread_count'].fillna(df['thread_count'].mean(), inplace=True)
```

```python
df['GPU_brand'] = df['GPU'].apply(lambda x: x.split()[0])
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
correlations = df.corr()['price'].abs().sort_values(ascending=False)
strong_features = correlations[correlations > 0.1].index.tolist()
if 'price' in strong_features:
    strong_features.remove('price')
X = df[strong_features]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

#  Predict on test set
y_pred = model.predict(X_test_scaled)


mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R² Score:", r2)
```
